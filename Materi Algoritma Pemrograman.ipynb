## Ekspresi Persamaan Matematika

### 1. Penjumlahan Sederhana

```python
# Input bilangan dari pengguna
bilangan1 = float(input("Masukkan bilangan pertama: "))
bilangan2 = float(input("Masukkan bilangan kedua: "))

# Hitung penjumlahan
hasil = bilangan1 + bilangan2

# Cetak hasil
print("Hasil penjumlahan:", hasil)
```

### 2. Pangkat (Eksponensial)

```python
# Input basis dan eksponen dari pengguna
basis = float(input("Masukkan basis: "))
eksponen = float(input("Masukkan eksponen: "))

# Hitung hasil pangkat
hasil = basis ** eksponen

# Cetak hasil
print("Hasil pangkat:", hasil)
```

### 3. Sisa Bagi (Modulo)

```python
# Input bilangan dari pengguna
bilangan1 = int(input("Masukkan bilangan pertama (integer): "))
bilangan2 = int(input("Masukkan bilangan kedua (integer): "))

# Hitung sisa bagi
hasil = bilangan1 % bilangan2

# Cetak hasil
print("Sisa bagi:", hasil)
```

### 4. Akar Kuadrat

```python
import math

# Input bilangan dari pengguna
bilangan = float(input("Masukkan bilangan: "))

# Hitung akar kuadrat
hasil = math.sqrt(bilangan)

# Cetak hasil
print("Akar kuadrat:", hasil)
```

### 5. Luas Segitiga

```python
# Input tinggi dan alas segitiga dari pengguna
tinggi = float(input("Masukkan tinggi segitiga: "))
alas = float(input("Masukkan alas segitiga: "))

# Hitung luas segitiga
luas = 0.5 * tinggi * alas

# Cetak hasil
print("Luas segitiga:", luas)
```

### 6. Konversi Satuan Panjang

```python
# Input panjang dalam meter dari pengguna
panjang_meter = float(input("Masukkan panjang dalam meter: "))

# Konversi panjang ke kilometer dan sentimeter
panjang_kilometer = panjang_meter / 1000
panjang_sentimeter = panjang_meter * 100

# Cetak hasil
print("Panjang dalam kilometer:", panjang_kilometer)
print("Panjang dalam sentimeter:", panjang_sentimeter)
```

### 7. Bilangan Faktorial

```python
# Input bilangan bulat positif dari pengguna
n = int(input("Masukkan bilangan bulat positif: "))

# Inisialisasi hasil dengan nilai awal 1
hasil = 1

# Hitung faktorial
for i in range(1, n + 1):
    hasil *= i

# Cetak hasil
print(f"Faktorial dari {n} adalah {hasil}")
```

## Percabangan If Else

#### Percabangan 'If' (Single Alternative Selection):

```python
umur = 20
if umur >= 18:
   print("Anda sudah dewasa.")
```

#### Percabangan 'If-Else' (Dual Alternative Selection):

```python
nilai = 80
if nilai >= 80:
    print("Anda lulus.")
else:
    print("Anda belum lulus.")
```

#### Percabangan 'If-Elif-Else' (Multiple Alternative Selection):

```python
nilai = 75
if nilai >= 90:
    print("A")
elif nilai >= 80:
    print("B")
elif nilai >= 70:
    print("C")
else:
    print("D")
```

#### Percabangan Bersarang (Nested If)

```python
# Menentukan apakah bilangan positif, negatif, atau nol, dan apakah bilangan tersebut merupakan bilangan genap atau ganjil

bil = float(input("Masukkan bilangan: "))

if bil > 0:
    # Bilangan positif
    if bil % 2 == 0:
        print("Bilangan positif dan genap")
    else:
        print("Bilangan positif dan ganjil")
elif bil < 0:
    # Bilangan negatif
    print("Bilangan negatif")
else:
    # Bilangan nol
    print("Bilangan nol")
```

## Nested If & Dasar Perluangan

#### If Nested
Kita juga dapat menggunakan pernyataan if di dalam pernyataan if. Hal ini dikenal sebagai pernyataan if bersarang.

Aturan sintaks dalam pernyataan if bersarang adalah:
*   Kita dapat menambahkan pernyataan else dan elif ke dalam pernyataan if sesuai kebutuhan
*   Kita juga dapat menyisipkan pernyataan inner if di dalam pernyataan else atau elif di luarnya (jika ada)
*   Kita dapat membuat beberapa lapisan pernyataan if.
 bersarang:
  








```python
# Contoh 1: Nested if dalam "Menentukan suatu angka positif, negatif atau 0"
number = int(input('Masukkan angka yang manu di cek: '))

# outer if statement
if (number >= 0):
    # inner if statement
    if number == 0:
        print('Angka ini 0')
    # inner else statement
    elif number == 7:
        print('Number is 7')
    else:
        print('Angka ini positive')
# outer else statement
else:
    print('Angka ini negative')
```

```python
# Contoh 2 : Saat kita menggunakan pernyataan Nested If, kita sering kali membuat perbandingan tambahan setelah pernyataan if biasa.
umur = 19
lulus = False
mendapat_SIM = True

# Lihat apakah orang tersebut berusia 18 tahun atau lebih
if umur >= 18:
    print('Anda berusia 18 tahun atau lebih. Selamat datang di masa dewasa!')
    if lulus:
        print('Selamat atas kelulusan Anda!')
    if mendapat_SIM:
        print('Selamat mengemudi!')
else:
    print("Anda belum berusia 18 tahun atau lebih. Selamat datang di masa kanak-kanak")
```

```python
# Contoh 3:
barang_dipesan = 16
barang_tersedia = 32
isi_paket = 8
print("Mendapatkan pesanan sebanyak", barang_dipesan, "barang. Tersedia di stok:", barang_tersedia)

# Membandingkan jumlah pesanan dengan stok
if barang_dipesan >= barang_tersedia:
    print("Tambahan persediaan diperlukan. Stok hampir habis!")
else:
    modulusPaket = barang_dipesan % isi_paket
    jumlahPaket = round(barang_dipesan/isi_paket)
    if isinstance(modulusPaket, int) and modulusPaket != 0:
        print(f'Kita memerlukan {round(jumlahPaket) + 1} paket untuk memenuhi pesanan ini!')
    elif isinstance(modulusPaket, int) and modulusPaket == 0:
        print(f'Kita memerlukan {jumlahPaket} paket untuk memenuhi pesanan ini!')
```

## Perulangan (for) - Counted loop

#### Perulangan (for) biasanya digunakan untuk mengulangi kode yang sudah diketahui banyak perulangannya.

Gambaran umum tentang bagaimana loop for bekerja:
*   Loop menginisialisasi iterator dengan item pertama dalam urutan
*   Ia mengeksekusi pernyataan dalam <command_expression> untuk nilai saat ini dari iterator.
*   Iterator diperbarui ke item berikutnya dalam urutan.
*   Langkah 2 dan 3 diulangi sampai akhir urutan tercapai, pada titik ini loop berakhir



#### Melakukan Pengulangan Beberapa Kali:

```python
for i in range(100, 0,-2):
    print(f"ini perulangan ke {i}")
```

### Perulangan Melalui Elemen dalam Daftar atau Larik:

```python
buah = ["apple", "pisang", "mangga"]
for buah in buah:
    print(f'Saya Suka Makan {buah}')
```

```python
buah = ["apple", "pisang", "mangga"]
for i in range(3):
    print(buah[i])
```

### Perulangan Melalui Karakter dalam String

```python
word = "Hello World"
for char in word:
    print(char, end="")
```

```python
angka = 5
for i in range(1, 6):
    print(angka, 'x', i, '=', angka*i)
```

## Perulangan (While) - Uncounted loop

##### Statement while sangat berkaitan dengan variabel boolean, atau logical statement. Karena penentuan kapan suatu blok kode akan diulang-ulang ditinjau dari True or False dari suatu pernyataan logika. Sehingga jika suatu kondisi itu selalu benar, maka perulangannya pun akan selalu di eksekusi.


#### Infinity Loop

```python
while (1 + 2 == 3):
    print('Halo dunia!')
```

#### Limited Loop

```python
# Contoh 1: Menampilkan angka urut dari 1 sampai 5 dengan While
i = 1

while i <= 5:
    print(i)
    i += 1
```

```python
# Contoh 2: menampilkan angka kelipatan 3 dari 1 sampai 100
i = 3
while i < 100:
    print(i)
    i = i + 3
```
